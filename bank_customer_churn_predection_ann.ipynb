```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
X=df.iloc[:,3:13]
y=df.iloc[:,13]
```

```python
geography=pd.get_dummies(X['Geography'],drop_first=True).astype(int)
gender=pd.get_dummies(X['Gender'],drop_first=True).astype(int)
```

```python
X=X.drop(['Geography','Gender'],axis=1)
```

```python
X=pd.concat([X,geography,gender],axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
#feature scalling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
 ## DL part, creating ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU, LeakyReLU
from tensorflow.keras.layers import Dropout
```

```python
#initialize the ANN
classifier = Sequential()

classifier.add(Dense(11, activation='relu', input_dim=11))
classifier.add(Dropout(0.3))

classifier.add(Dense(7, activation='relu'))
classifier.add(Dropout(0.3))

classifier.add(Dense(6, activation='relu'))

classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# early stopping
import tensorflow as tf
early_stopping= tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=early_stopping)
```

```python
# predection

y_pred = classifier.predict(X_test)
y_pred=(y_pred>=0.5)
```

```python
from sklearn.metrics import confusion_matrix #confusion matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```
